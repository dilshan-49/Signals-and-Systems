```python
import numpy as np      #importing numpy library
import matplotlib.pyplot as plt     #importing matplotlib library
%matplotlib inline
```

Let's plot $x(t)= ACos(\omega_{0}t+\phi)$

```python
fs=44100 #sampling frequency
ts=1/fs #sampling time
t=np.arange(-1.,1.,ts) #time for 2 seconds
A=1.0 #Amplitude
f=2.0 #Frequency

x=A*np.cos(2*np.pi*f*t) #Signal

fig,ax=plt.subplots()
ax.plot(t,x)

ax.set(xlabel='Time(s)',ylabel='Amplitude')     #setting x and y labels
ax.minorticks_on()
ax.grid(True)
ax.grid(visible=True, which='minor', color='#999999', linestyle='-', alpha=0.3) #enabling minor grid
ax.title.set_text(r'$x(t)= ACos(\omega_{0}*t+\phi)$')   #setting title of the plot
plt.show()
```

#### Plot the following Graphs
1.    $x_{1}(t)=cos(2\pi t)$
2.    $x_{2}(t)=0.5cos(2\pi t)$
3.    $x_{3}(t)=-sin(2\pi t)$
4.    $x_{4}(t)=cos(4\pi t)$

```python
fs = 44100 # 44,100Hz sampling frequency
ts = 1/fs
t = np.arange(-1., 1., ts) # A linearlyspaced array with step ts
f = 2 # 2 Hz
fig, axes = plt.subplots(2,2, sharex='all', sharey='all', figsize=(18,18))

#####################  Signals  #####################

x1=np.cos(2*np.pi*t) 
x2=0.5*np.cos(2*np.pi*t) 
x3=-np.sin(2*np.pi*t) 
x4=np.cos(4*np.pi*t)

######################################################

#plotting x1
axes[0,0].plot(t, x1)
axes[0,0].set_title(r'$x(t)=Cos(2\pi t)$')              #setting title of the plot
axes[0,0].set(xlabel='Time(s)',ylabel='$x_{1}(t)$')     #setting x and y labels
axes[0,0].minorticks_on()
axes[0,0].grid(visible=True, which='both')              #enabling grid

#plotting x2
axes[0,1].plot(t, x2) 
axes[0,1].set_title(r'$x(t)=0.5Cos(2\pi t)$')           #setting title of the plot
axes[0,1].set(xlabel='Time(s)',ylabel='$x_{2}(t)$')     #setting x and y labels
axes[0,1].minorticks_on()
axes[0,1].grid(visible=True, which='both')              #enabling grid

#plotting x3
axes[1,0].plot(t, x3)
axes[1,0].set_title(r'$x(t)=-Sin(2\pi t)$')             #setting title of the plot
axes[1,0].set(xlabel='Time(s)',ylabel='$x_{3}(t)$')     #setting x and y labels
axes[1,0].minorticks_on()
axes[1,0].grid(visible=True, which='both')              #enabling grid

#plotting x4
axes[1,1].plot(t, x4) 
axes[1,1].set_title(r'$x(t)=Cos(4\pi t)$')              #setting title of the plot
axes[1,1].set(xlabel='Time(s)',ylabel='$x_{4}(t)$')     #setting x and y labels
axes[1,1].minorticks_on()
axes[1,1].grid(visible=True, which='both')              #enabling grid


plt.show()
```

#### Plotting 2 waves in a same plot


```python
fs = 4.4e4                  # 44KHz sampling frequency
ts = 1/fs
t = np.arange(-1., 1., ts)  # A linearlyspaced array with step ts
f = 2                       # 2 Hz
w0=2*np.pi*f                #angular frequency

phi=0                       #phase
x1=np.cos(w0*t+phi)         #Signal-1

phi=np.pi/2                 
x2=np.cos(w0*t+phi)         #Signal-2


fig,ax=plt.subplots()
ax.plot(t,x1,label=r'$x(t)=Cos(\omega_{0}*t)$')         #plotting x1
ax.plot(t,x2,label=r'$x(t)=Cos(\omega_{0}*t+\pi/2)$')   #plotting x2
ax.set(xlabel='$t$',ylabel='$x(t)$')                    #setting x and y labels
ax.minorticks_on()
ax.grid(visible=True, which='both')                     #enabling grid

ax.set_title(r'$x(t)=ACos(\omega_{0}t+\phi)\,  with\, \phi=0 and \, \phi=\pi/2$')

plt.show()
```

#### Real CT Exponential
$\qquad x(t)=Ce^{at} $

```python
fs=4.4e4 # 44KHz sampling frequency
ts=1/fs #sampling time
t=np.arange(-1.,1.,ts)

C=1
a=1.2
xt=C*np.exp(a*t) #Signal

fig,ax=plt.subplots()
ax.plot(t,xt)
ax.set(xlabel='Time(s)',ylabel='x(t)')     #setting x and y labels
ax.minorticks_on()
ax.grid(visible=True, which='both')         #enabling grid
ax.title.set_text(r'$x(t)=Ce^{at}$')       #setting title of the plot

plt.show()
```
